```python
import ipywidgets as widgets
from IPython.display import display
from plotly.offline import iplot
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
import pandas as pd
import requests
#Cars URL
url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)
data_df = pd.DataFrame(data['result']['records'])

# Tab 1: Data Overview
tab1_content = widgets.Output()
with tab1_content:
    print("Statistics for all Data:")
    fig = px.scatter(
    data_df,
    x='shnat_yitzur',        # Year of manufacture
    y='kvutzat_zihum',       # Pollution group or similar
    color='tozeret_nm',      # Manufacturer
    hover_name='degem_nm',   # Model
    size_max=60)
fig.show()

# Tab 2: Raw Data
tab2_content = widgets.Output()
with tab2_content:
    print("Data Overview in Table:")
    pd.set_option("display.max_columns", None)
    #יציג את 5 השורות הראשונות של הנתונים
    display(data_df)

# Tab 3: Charts
tab3_content = widgets.Output()
with tab3_content:
    print("גרף - מספר המכוניות השונות שיוצרו בכל שנה:")

    # Create the figure
    fig, ax = plt.subplots(figsize=(14, 6))

    # Plot: Count of cars by year
    sns.countplot(data=data_df, x='shnat_yitzur', order=sorted(data_df['shnat_yitzur'].dropna().unique()), ax=ax)

    # Style
    ax.set_title('מספר המכוניות השונות שיוצרו בכל שנה', fontsize=14)
    ax.set_xlabel('שנת ייצור', fontsize=12)
    ax.set_ylabel('מספר מכוניות', fontsize=12)
    plt.xticks(rotation=45)

    # Show
    display(fig)
    plt.close(fig)

# Create Tabs
tabs = widgets.Tab(children=[tab1_content, tab2_content,tab3_content])
tabs.set_title(0, 'Tab 1: Statistics')
tabs.set_title(1, 'Tab 2: Table')
tabs.set_title(2, 'Tab 3: Cars per year')



# Display Tabs
display(tabs)
```
